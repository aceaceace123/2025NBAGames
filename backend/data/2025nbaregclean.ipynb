# function

```python
def avg_stats(team1, team2, df_):
    stats_ = df_[
        (df_['MATCHUP'] == f'{team1} vs. {team2}') |
        (df_['MATCHUP'] == f'{team2} vs. {team1}') |
        (df_['MATCHUP'] == f'{team1} @ {team2}') |
        (df_['MATCHUP'] == f'{team2} @ {team1}')
    ][['TEAM_ABBREVIATION', 'MATCHUP', 'WL', 'PTS', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'FG_PCT', 'FG3_PCT', 'FT_PCT']]
    
    print('num of matches:', int(stats_.shape[0] / 2))

    # ÂãùÂ†¥Êï∏Áµ±Ë®à
    win_df = stats_.groupby(['TEAM_ABBREVIATION', "WL"]).size().unstack(fill_value=0)[['W']]
    win_df = win_df.rename(columns={"W": "Wins"})

    # Âπ≥ÂùáÊï∏ÊìöÔºàÂü∫Êú¨Ê¨Ñ‰ΩçÔºâ
    avg_df = stats_.groupby(['TEAM_ABBREVIATION'])[
        ['PTS', 'REB', 'AST', 'STL', 'BLK', 'TOV']
    ].mean().round(1)

    # ÂëΩ‰∏≠ÁéáÊ¨Ñ‰ΩçÔºö‰πò 100 ÂÜçÂõõÊç®‰∫îÂÖ•
    pct_df = stats_.groupby(['TEAM_ABBREVIATION'])[
        ['FG_PCT', 'FG3_PCT', 'FT_PCT']
    ].mean().mul(100).round(1)

    # Âêà‰ΩµÂëΩ‰∏≠ÁéáÊ¨Ñ‰ΩçÂà∞ avg_df ÂæåÊñπ
    avg_df = pd.concat([avg_df, pct_df], axis=1)

    # ÊèíÂÖ•ÂãùÂ†¥Êï∏Âú®ÊúÄÂâçÈù¢
    avg_df.insert(0, 'Wins', win_df['Wins'])

    return avg_df
```

```python
import json

def save_all_pairwise_matchups(df_):
    teams = ['OKC','HOU','LAL','DEN','LAC','MIN','GSW','MEM','CLE','BOS','NYK','IND','MIL','DET','ORL','MIA']
    all_data = {}

    for i in range(len(teams)):
        team1 = teams[i]
        for j in range(i+1, len(teams)):
            team2 = teams[j]
            try:
                # ÂèñÂæóÂπ≥ÂùáÁµ±Ë®àË≥áÊñô
                avg_df = avg_stats(team1, team2, df_).reset_index()

                # ÂëΩ‰∏≠ÁéáÊ¨Ñ‰ΩçÈáçÂëΩÂêç
                rename_dict = {
                    "FG_PCT": "FG%",
                    "FG3_PCT": "3P%",
                    "FT_PCT": "FT%"
                }
                avg_df = avg_df.rename(columns=rename_dict)

                # ÁµÑÂêà key
                matchup_key = 'vs'.join(avg_df['TEAM_ABBREVIATION'].tolist())

                # Âª∫Á´ã {Ê¨Ñ‰Ωç: [ÂÄº1, ÂÄº2]} ÁµêÊßã
                matchup_data = {
                    col: avg_df[col].tolist()
                    for col in avg_df.columns if col != 'TEAM_ABBREVIATION'
                }

                # Âä†ÂÖ•Âà∞Á∏ΩË≥áÊñô‰∏≠
                all_data[matchup_key] = matchup_data
                print(f"‚úÖ Â∑≤Âä†ÂÖ•Â∞çÊà∞Ôºö{matchup_key}")

            except Exception as e:
                print(f"‚ö†Ô∏è Ë∑≥ÈÅé {team1} vs {team2}Ôºö{e}")

    # ÂÑ≤Â≠òÁÇ∫ JSON
    with open("/Users/liang/Desktop/2025NBAGames/backend/data/matchup_stats.json", "w", encoding="utf-8") as f:
        json.dump(all_data, f, ensure_ascii=False, indent=4)

    print("üéâ ÊâÄÊúâÈöä‰ºçÂ∞çÊà∞Ë≥áÊñôÂ∑≤ÊàêÂäüÂÑ≤Â≠òÂà∞ matchup_stats.json")
```

```python
from bs4 import BeautifulSoup as bs
import requests

# Step 1: ÂèñÂæó team names
url1 = 'https://www.nba.com/teams'
r1 = requests.get(url1, headers={"User-Agent": "Mozilla/5.0"})
soup1 = bs(r1.content, "html.parser")
teams_name_tags = soup1.find_all(class_="Anchor_anchor__cSc3P TeamFigure_tfMainLink__OPLFu")
team_names = [tag.text.strip() for tag in teams_name_tags]

# Step 2: ÂèñÂæó team ids
url2 = 'https://www.nba.com/stats/teams'
r2 = requests.get(url2, headers={"User-Agent": "Mozilla/5.0"})
soup2 = bs(r2.content, "html.parser")
teams_info_tags = soup2.find_all(class_="Anchor_anchor__cSc3P StatsTeamsList_teamLink__q_miK")
team_ids = [tag.get("href").split("/")[-1] for tag in teams_info_tags]

# Step 3: ÈÖçÂ∞çÊàê dict
team_dict = dict(zip(team_ids, team_names))

# Step 4: È°ØÁ§∫ÊàñÂÑ≤Â≠ò
print(team_dict)
```

```python
url='https://www.nba.com/stats/teams'
r=requests.get(url)
contents=bs(r.content)
teams_info=contents.find_all(class_="Anchor_anchor__cSc3P StatsTeamsList_teamLink__q_miK")
all_rosters = []
# Êì∑Âèñ href Â±¨ÊÄßÔºàÂç≥Ë∂ÖÈÄ£ÁµêÔºâ
for a in teams_info:
    link = a.get("href")
    # print(link)
    team_id = link.split("/")[-1]
    url = f"https://stats.nba.com/stats/commonteamroster?LeagueID=00&Season=2024-25&TeamID={team_id}"
    headers = {
        "User-Agent": "Mozilla/5.0",
        "Referer": "https://www.nba.com/",
        "x-nba-stats-origin": "stats",
        "x-nba-stats-token": "true"
    }

    res = requests.get(url, headers=headers)
    data = res.json()

    # ÊäìÂèñÁ¨¨‰∏ÄÁµÑÁµêÊûúÔºàÈÄöÂ∏∏ÊòØÁ∏ΩË¶ΩË≥áÊñôÔºâ
    headers = data['resultSets'][0]['headers']
    rows = data['resultSets'][0]['rowSet']

    df = pd.DataFrame(rows, columns=headers)
    df["TEAM_NAME"] = new_dict[team_id]
    all_rosters.append(df)

full_roster_df = pd.concat(all_rosters, ignore_index=True)
    
```

```python
headers = data['resultSets'][0]['headers']
rows = data['resultSets'][0]['rowSet']
df = pd.DataFrame(rows, columns=headers)
print(df.columns)
print(df.shape)
```
